```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")

df = pd.read_csv('Atividade.csv')

display(df.head())
df.info()

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Sales', bins=30, kde=True, color='lime')
plt.title('Distribuio do Valor de Vendas')
plt.xlabel('Vendas')
plt.ylabel('Contagem')
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Category', y='Profit', palette='Set2')
plt.title('Lucro por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Lucro')
plt.show()

plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Sales', hue='Region', fill=True, palette='crest')
plt.title('Vendas por Regio')
plt.xlabel('Vendas')
plt.show()


plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Sales', hue='Segment', multiple='stack', bins=25, palette='pastel')
plt.title('Distribuio de Vendas por Segmento')
plt.xlabel('Vendas')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Category', y='Profit', hue='Region', palette='Set3')
plt.title('Lucro por Categoria Separado por Regio')
plt.xlabel('Categoria')
plt.ylabel('Lucro')
plt.legend(title='Regio')
plt.show()

plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Category', y='Quantity', hue='Segment', split=True, palette='muted')
plt.title('Distribuio da Quantidade por Categoria e Segmento')
plt.xlabel('Categoria')
plt.ylabel('Quantidade')
plt.show()
```
