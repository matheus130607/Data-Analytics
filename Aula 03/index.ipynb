```python
# Para rodar a aplicaÃ§Ã£o
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#ConfiguraÃ§Ã£o Visual
sns.set_theme(style="darkgrid")

#Carregar nabase de dados
#Como o arquivo estÃ¡ na mesma pasta, basta colocar o nome
df = pd.read_csv('titanic.csv')

#Mostra as primeiras 5 linhas formadas
display(df.head())
#DICA: No VS Code, use 'display()' ao inves de 'print()' para tabelas bonitas

#Vefirficando tipos e dados e nulos
df.info()

#AnÃ¡lise de Idade
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', bins=30, kde=True, color='lime')
plt.title('DistribuiÃ§Ã£o de Idade dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Contagem')
plt.show()

#ComparaÃ§Ã£o de Tarifa por Classe
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Pclass', y='Fare', palette='Set2')
plt.title('PreÃ§o da Passagem por Classe (Bilhetes)')
plt.show()

#Apenas a linha de desnidade (sem barras)
plt.figure(figsize=(10,6))
sns.kdeplot(data=df, x='Age', fill=True, common_norm=False, palette='crest')
plt.title('Densidade de Idade: Quem Sobriviveu (1) vs Morreu (0)')
plt.show()

#Histograma de Idade separado por Sexo
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Age', hue='Sex', multiple='stack', bins=25, palette='pastel')
plt.title('DistribuiÃ§Ã£o de Idade: Homens vs Mulheres')
plt.xlabel('Idade')
plt.show()

#PreÃ§o x Classe, separado por SobrevivÃªncia
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Pclass', y='Fare', hue='2urvived', palette='Set3')
plt.title('Impacto do PreÃ§o na SobrevivÃªncia po Classe')
plt.xlabel('Classe')
plt.ylabel('PreÃ§o da Passagem')
plt.legend(title='Sobreviveu?')
plt.show()

#Violin Plot: A evoluÃ§Ã£o do Bonx-plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Pclass', y='Age', split=True, hue='Sex', palette='muted')
plt.title('DistribuiÃ§Ã£o de Idade por Classe e Sexo')
plt.show


```
